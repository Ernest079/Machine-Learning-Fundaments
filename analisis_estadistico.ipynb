```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Cargar Datos
partidos_cebollitas = pd.read_csv('partidos_cebollitas.csv')
partidos_cebollitas.head()
```

```python
partidos_cebollitas.describe()
```

```python
# La función .mean() nos da el promedio de goles que hace Cebollitas FC como local y como visitante

goles_local = partidos_cebollitas[partidos_cebollitas['equipo_local'] == 'Cebollitas FC']['goles_local'].mean()
goles_visitante = partidos_cebollitas[partidos_cebollitas['equipo_visitante'] == 'Cebollitas FC']['goles_visitante'].mean()

print(f'Goles promedio como local: {goles_local}')
print(f'Goles promedio como visitante: {goles_visitante}')  
```

```python
#El histograna nos permite ver la distribución de los datos de goles que hace Cebollitas FC
#Seaborn (sns) es una libreria util para hacer visualizaciones estadisticas

fig, ax = plt.subplots(1, 2, figsize=(12, 4))

sns.histplot(partidos_cebollitas['goles_local'], bins=range(0, 10), kde=True, ax=ax[0], discrete=True)
ax[0].set_title('Distribución de Goles como Local')

sns.histplot(partidos_cebollitas['goles_visitante'], bins=range(0, 10), kde=True, ax=ax[1], discrete=True, color='orange')
ax[1].set_title('Distribución de Goles como Visitante')
```

# Distribuciones visuales mediante histogramas permiten descubrir:

# Frecuencias altas para resultados de dos, tres o cuatro goles jugando en casa.
# Muchos partidos sin goles y varios de alta anotación (hasta cinco goles) jugando como visitante.

```python
#Boxplot nos permite ver la distribución de los datos y detectar posibles valores atípicos (outliers)

sns.boxplot(data=partidos_cebollitas[['goles_local', 'goles_visitante']])
plt.title('Boxplot de Goles como Local y Visitante')
```

```python
#scatterplot nos permite ver la relación entre dos variables, en este caso la posesión y los goles como local

sns.scatterplot(x='posesion_local (%)', y='goles_local', data=partidos_cebollitas)
plt.title('Relación entre Posesión y Goles como Local')
```

```python
#Mapa de calor (heatmap) para visualizar correlaciones entre variables

plt.figure(figsize=(10, 6))
sns.heatmap(partidos_cebollitas[[
  'goles_local', 'goles_visitante', 'posesion_local (%)', 'posesion_visitante (%)', 'tiros_arco_local', 'tiros_arco_visitante'
]].corr(), annot=True, cmap='coolwarm')
plt.title('Mapa de Calor de Correlaciones')
```
